```python
import numpy as np

# Filing missing values. This confirms that the absence of a weapon is the baseline.
df_v3['Weapon Desc'].fillna('None', inplace=True)

# A function that groups the 79 weapon types into a few categories.
def categorize_weapon(description):
    # Handle potential missing values (nan) first
    if not isinstance(description, str):
        return 'None'

    # Standardize to uppercase for easier matching
    desc_upper = description.upper()

    # --- Category Checks ---

    # 1. Firearms (including simulated/toy guns used as threats)
    firearm_keywords = ['GUN', 'PISTOL', 'REVOLVER', 'RIFLE', 'FIREARM', 'SHOTGUN',
                        'ASSAULT WEAPON', 'UZI', 'AK47', 'MAC-10', 'MAC-11', 'SUB-MACHINE']
    if any(keyword in desc_upper for keyword in firearm_keywords):
        return 'Firearm'

    # 2. Edged Weapons
    edged_keywords = ['KNIFE', 'BLADE', 'RAZOR', 'CUTTING INSTRUMENT', 'MACHETE',
                      'CLEAVER', 'SWORD', 'ICE PICK', 'DIRK/DAGGER', 'SCISSORS']
    if any(keyword in desc_upper for keyword in edged_keywords):
        return 'Edged Weapon'

    # 3. Bodily Force
    force_keywords = ['STRONG-ARM', 'HANDS', 'FIST', 'FEET', 'BODILY FORCE', 'PHYSICAL PRESENCE']
    if any(keyword in desc_upper for keyword in force_keywords):
        return 'Bodily Force'

    # 4. Threats (Verbal or Written)
    threat_keywords = ['VERBAL THREAT', 'BOMB THREAT', 'DEMAND NOTE']
    if any(keyword in desc_upper for keyword in threat_keywords):
        return 'Threat'

    # 5. Unknown Weapons (check this after specific types)
    if 'UNKNOWN' in desc_upper:
        return 'Unknown'

    # 6. None (for explicitly stated no-weapon cases)
    if 'NONE' in desc_upper:
        return 'None'

    # 7. If none of the above, classify as Other Weapon
    return 'Other Weapon'

# Create the new column by applying the function
df_v3['Weapon_Category'] = df_v3['Weapon Desc'].apply(categorize_weapon)

# Verify the results
print(df_v3['Weapon_Category'].value_counts())
df_v3
```

```python
# Creating the column 'involved_physical_force' from certain mocodes

# First, ensure your 'Mocodes_List' column exists and is clean
df_v4['Mocodes'].fillna('', inplace=True)
df_v4['Mocodes_List'] = df_v4['Mocodes'].str.split()

# Step 1: Define the list of codes that represent physical force
force_codes = [
    '0356', '0400', '0401', '0406', '0407', '0408', '0413',
    '0416', '0417', '0419', '0426', '0429', '0430', '0444',
    '0446', '0447', '0448', '1281', '1601'
]

# Step 2: Write a function to check if any of these codes exist in a list
def check_for_codes(code_list, target_codes):
    # The set intersection is a fast way to check for common elements
    return 1 if set(code_list) & set(target_codes) else 0

# Step 3: Apply this function to create your new feature column
df_v4['involved_physical_force'] = df_v4['Mocodes_List'].apply(lambda x: check_for_codes(x, force_codes))

# Step 4: Verify the result
print(df_v4['involved_physical_force'].value_counts())
```

```python
# Creating the column 'involved_victim_restraint' from certain mocodes
# Re-usable function from the previous example
def check_for_codes(code_list, target_codes):
    """Checks if any target codes are present in a list of codes."""
    return 1 if set(code_list) & set(target_codes) else 0

# Step 1: Define the list of codes that represent victim restraint
restraint_codes = [
    '0341', # Makes vict lie down
    '0402', # Blindfolded
    '0405', # Bound
    '0409', # Cover mouth w/hands
    '0411', # Covered victim's face
    '0414', # Gagged
    '0415', # Handcuffed/Metal
    '0423', # Tied victim to object
    '0435', # Chain
    '0436', # Clothing
    '0437', # Flexcuffs/Plastic Tie
    '0438', # Rope/Cordage
    '0439', # Tape/Electrical etc...
    '0440', # Telephone/Electric Cord
    '0441'  # Wire
]

# Step 2: Apply the function to create your new feature column
df_v4['involved_victim_restraint'] = df_v4['Mocodes_List'].apply(lambda x: check_for_codes(x, restraint_codes))

# Step 3: Verify the result
print("--- 'involved_victim_restraint' Distribution ---")
print(df_v4['involved_victim_restraint'].value_counts())
```

```python
# Creating the column 'showed_planning_or_stealth' from certain mocodes
# Step 1: Define the specific codes for planning and stalking
planning_codes = [
    '0313', # Hid in building
    '0335', # Cases location
    '0338', # Disables Telephone
    '0339', # Disables video camera
    '0340', # Suspect follows victim/follows victim home
    '0347'  # Stalks vict
]

# Step 2: Write a more specific function for this feature
def check_for_planning(code_list):
    """
    Checks for specific planning codes OR any disguise code (starting with '02').
    """
    # Check for specific planning codes
    if set(code_list) & set(planning_codes):
        return 1

    # Check for any disguise code (those in the 0200 series)
    for code in code_list:
        if code.startswith('02'):
            return 1

    return 0

# Step 3: Apply this new function to create the feature column
df_v4['showed_planning_or_stealth'] = df_v4['Mocodes_List'].apply(check_for_planning)

# Step 4: Verify the result
print("\n--- 'showed_planning_or_stealth' Distribution ---")
print(df_v4['showed_planning_or_stealth'].value_counts())
```

## Incident Identifiers & Timestamps
DR_NO: The unique report number that identifies a single crime incident.

DATE OCC: The date the crime actually occurred. This is the primary column for creating your victim timeline.

Date Rptd: The date the crime was reported to the police.

TIME OCC: The time the crime occurred, in 24-hour military time format.

## Crime Descriptors
Crm Cd: The numerical code for the specific type of crime.

Crm Cd Desc: The text description of the crime (e.g., "BURGLARY FROM VEHICLE").

Part 1-2: A classification of the crime's severity, where Part 1 crimes are generally more serious.

Weapon_Category: Your engineered feature describing the type of weapon used (e.g., 'Firearm', 'Bodily Force', 'None').

## Location Information
AREA: The numerical code for the police reporting district.

AREA NAME: The name of the police reporting district (e.g., "Central").

Premis Cd: The numerical code for the type of premise where the crime occurred.

Premis Desc: The text description of the premise (e.g., "STREET", "SINGLE FAMILY DWELLING").

LOCATION: The specific street address or intersection of the crime.

LAT: The latitude coordinate for the crime location.

LON: The longitude coordinate for the crime location.

## Victim Demographics
Vict Age: The age of the victim.

Vict_Sex_Clean: Your cleaned feature for the victim's sex ('Male', 'Female', 'Unknown/Other').

Vict_Descent_Clean: Your cleaned feature for the victim's race/ethnicity.

## Engineered Behavioral Features
These are the new features we created from the Mocodes column.

involved_physical_force: A binary (1/0) feature indicating if the crime involved direct physical violence against the victim.

involved_victim_restraint: A binary (1/0) feature indicating if the victim was bound, held, or otherwise restrained.

showed_planning_or_stealth: A binary (1/0) feature indicating if the suspect used a disguise, stalked the victim, or showed other signs of premeditation.

# Creating the Dependent Variable (Target) ("Is_Violent_Crime")
* 1 = Crime was violent
* 0 = Crime was non-violent



```python
# UCR-aligned keywords to search for in the 'Crm Cd Desc' column.
UCR_VIOLENT_KEYWORDS = [
    # For "murder and nonnegligent manslaughter"
    'HOMICIDE',
    'MANSLAUGHTER',

    # For "rape"
    'RAPE',

    # For "robbery"
    'ROBBERY',

    # For "aggravated assault"
    'AGGRAVATED ASSAULT',
    'ADW' # Often stands for "Assault with a Deadly Weapon"
]

# This list remains the same, as it directly addresses the "involve force or threat of force" principle.
VIOLENT_WEAPON_CATEGORIES = [
    'Firearm', 'Edged Weapon', 'Bodily Force', 'Other Weapon'
]
```

```python
def classify_violence_ucr(row):
    # Condition 1: Check if the crime involved force or threat of force (via weapon).
    if row['Weapon_Category'] in VIOLENT_WEAPON_CATEGORIES:
        return 1

    # Condition 2: Check if the crime is one of the four named UCR offenses.
    crime_desc = str(row['Crm Cd Desc']).upper()
    if any(keyword in crime_desc for keyword in UCR_VIOLENT_KEYWORDS):
        return 1

    # If neither condition is met, classify as non-violent.
    return 0
```

```python
# Apply the new, UCR-aligned function to each row.
repeat_victims_df['Is_Violent_Crime'] = repeat_victims_df.apply(classify_violence_ucr, axis=1)

# Verify the new distribution.
print("New Distribution of the 'Is_Violent_Crime' Target Variable:")
print(repeat_victims_df['Is_Violent_Crime'].value_counts(normalize=True))
```
